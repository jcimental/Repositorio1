```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def escalon(a):
  return np.float32(a>=0)

def mse(y, t):
  return 0.5*np.sum((t-y)**2)/len(y)

x = np.array([[0,0], [0,1], [1,0], [1,1]])
t = np.array([0, 1, 0, 1])

w = np.random.randn(2)
b = np.random.randn(1)

epochs = 200
alpha = 0.01
epoch_error = []

for epoch in range(epochs):
  y = np.zeros_like(t)

  for i in range(len(x)):
    a = np.dot(x[i], w) - b
    y[i] = escalon(a)
    d_w = -(t[i]-y[i])*x[i]
    d_b = t[i]-y[i]
    w -= alpha*d_w
    b -= alpha*d_b

  error = mse(y, t)
  epoch_error.append(error)

  if error == 0:
    break

rango_x = np.arange(1, len(epoch_error) + 1)
x0 = np.linspace(-0.5, 2, 5)
x1 = (b-x0*w[0])/w[1]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(rango_x, epoch_error)
plt.plot('perceptron learning rule')
plt.xlabel('epochs')
plt.ylabel('MSE')
plt.show()
```

```python
plt.scatter(x[:,0], x[:,1], c=t)
plt.plot(x0, x1)
plt.title('Decision boundary')
plt.xlabel('XO')
plt.ylabel('X1')
plt.show()
```

```python
#### DATOS ARTIFICIALES

samples = 1000
negative_samples = np.random.multivariate_normal(mean=[0,3], cov=[[1,0.5],[0.5,1]], size=samples)

positive_samples = np.random.multivariate_normal(mean=[3,0], cov=[[1,0.5],[0.5,1]], size=samples)

inputs = np.vstack((negative_samples, positive_samples))
labels = np.vstack((np.zeros((samples,1)), np.ones((samples,1))))

print('Inputs shapes', inputs.shape)
print('Inputs shapes', labels.shape)

plt.figure(figsize=(10, 5))
plt.scatter(inputs[:,0], inputs[:,1], c=labels[:,0])
plt.title('Patrón de entrada')
plt.xlabel('XO')
plt.ylabel('X1')
plt.show()
```

```python
class LogisticRegression():
  def __init__(self, epochs =200, learning_rate=0.001, batch_size=256):
    self.epochs = epochs
    self.learning_rate = learning_rate
    self.batch_size = batch_size
    self.batch_error = []
    self.W = np.random.randn(2,1)
    self.b = np.random.randn(1)

  def cross_entropy(self, T, Y):
    return -np.mean(T*np.log(Y) + (1-T)*np.log(1-Y))

  def sigmoid(self, A):
    return 1 / (1 + np.exp(-A))

  def predict(self, inputs):
    return np.dot(inputs, self.W) - self.b

  def fit(self, X, T):
    n_samples = np.shape(X)[0]                            #### 2000 datos

    for epoch in range(self.epochs):
      num_radom = np.random.permutation(n_samples)
      X_shuffle = X[num_radom]
      T_shuffle = T[num_radom]

      for i in range(0, n_samples, self.batch_size):
        X_batch = X_shuffle[i:i+self.batch_size]
        T_batch = T_shuffle[i:i+self.batch_size]

        A = self.predict(X_batch)
        Y = self.sigmoid(A)

        d_w = -(T_batch - Y)*X_batch
        d_b = T_batch - Y

        d_w = np.mean(d_w, axis=0).reshape((np.shape(self.W)))
        d_b = np.mean(d_b, axis=0).reshape((np.shape(self.b)))

        self.W -= self.learning_rate*d_w
        self.b -= self.learning_rate*d_b

      Y_pred = self.predict(X)
      Y_sig = self.sigmoid(Y_pred)
      error = self.cross_entropy(T, Y_sig)
      self.batch_error.append(error)

  @property
  def weights(self):
    return [self.W, self.b]

  @property
  def epoch(self):
    return self.batch_error
```

```python
model = LogisticRegression(epochs=200, learning_rate=0.01, batch_size=256)
model.fit(inputs, labels)
W, b = model.weights
batch_error = model.batch_error
epoch_vector = np.arange(len(batch_error))
##epoch_error = np.arrange(batch_error)
X0 = np.linspace(-2, 6, 2)
X1 = (b -X0*W[0])/W[1]

plt.figure(figsize=(10, 5))
plt.scatter(inputs[:,0], inputs[:,1], c=labels[:,0])
plt.plot(X0, X1)
plt.title('Patrón de entrada')
plt.xlabel('XO')
plt.ylabel('X1')
plt.show()
print()

plt.figure(figsize=(10, 5))
plt.plot(epoch_vector, batch_error)
plt.title('Error de entrenamiento')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.show()
```
